```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("PSL.csv")
```

```python
df.head()
```

# Missing Values Check 

```python
df.isnull().sum()
```

```python
df['extras_type']=df['extras_type'].fillna('Unknown')
df['dismissal_kind']=df['dismissal_kind'].fillna('Not Dismissed')
df['fielder']=df['fielder'].fillna('Not Available')
df['player_dismissed']=df['player_dismissed'].fillna('Not Available')
df['player_of_match']=df['player_of_match'].fillna('Unknown')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.head()
```

```python
df.drop('team_win', axis=1, inplace=True)
```

```python
df.head()
```

# Which team has won the most matches?

```python
team_wins = df['winner'].value_counts()
```

```python
print(team_wins)
```

# Visualization:


```python
team_wins.plot(kind='bar', color='skyblue')
plt.title('Most Matches Won by Teams')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```

# Conclusion

Islamabad United has won the most matches in PSL history

# Top 10 run scorers (batter-wise)

```python
top_scorers = df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
print(top_scorers)
```

# Visulization

```python
top_scorers.plot(kind='barh', color='orange')
plt.title('Top 10 Run Scorers')
plt.xlabel('Total Runs')
plt.ylabel('Batter')
plt.show()
```

# Most Wickets Taken

```python
wickets = df[df['is_wicket'] == True]
top_bowlers = wickets['bowler'].value_counts().head(10)
print(top_bowlers)
```

# Visualizatiom

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette='viridis')
plt.title('Top 10 Wicket Takers in PSL', fontsize=14)
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.tight_layout()
plt.show()
```

# Most "Man of the Match" Awards

```python
mom_counts = df['player_of_match'].value_counts().head(10)
```

```python


plt.figure(figsize=(8, 8))
plt.pie(mom_counts, labels=mom_counts.index,startangle=140, autopct='%1.1f%%',  colors=plt.cm.tab10.colors)
plt.title('Top 10 Players with Most "Man of the Match" Awards')
plt.legend(mom_counts.index, title="Players", loc="best", bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

#  Average Score per Season / Team

```python
# Batting score per inning
avg_score_season = df.groupby(['season', 'batting_team'])['total_runs'].mean().unstack()
avg_score_season.plot(kind='bar', figsize=(14, 6))
plt.title('Average Score per Season per Team')
plt.xlabel('Season')
plt.ylabel('Average Runs per Inning')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

match_totals = df.groupby(['season', 'batting_team', 'match_id'])['total_runs'].sum().reset_index()


avg_score_season = match_totals.groupby(['season', 'batting_team'])['total_runs'].mean().unstack()

avg_score_season.plot(kind='bar', figsize=(14, 6))
plt.title('Average Score per Season per Team')
plt.xlabel('Season')
plt.ylabel('Average Runs per Inning')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Most Sixes or fours 

```python
#for Sixes
sixes = df[df['batsman_runs'] == 6].groupby('batter').size().sort_values(ascending=False).head(10)
sixes.plot(kind='barh', color='purple')
plt.title('Top 10 Players with Most Sixes')
plt.xlabel('Number of Sixes')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Most Fours

fours = df[df['batsman_runs'] == 4].groupby('batter').size().sort_values(ascending=False).head(10)
fours.plot(kind='bar', color='red')
plt.title('Top 10 Players with Most Fours')
plt.xlabel('Number of Fours')
# plt.gca().invert_yaxis()
plt.show()
```

# Venue-wise Performance (Average Runs)

```python
venue_scores = df.groupby('venue')['total_runs'].mean().sort_values(ascending=False)
venue_scores.plot(kind='barh', color='teal', figsize=(10,6))
plt.title('Average Total Runs by Venue')
plt.xlabel('Average Runs')
plt.gca().invert_yaxis()
plt.show()
```
